```python
from tabula import read_pdf
import tabula
import tabulate
import camelot
```

```python
pdfData = read_pdf('table.pdf')
pdfData
```

```python
pdfData = read_pdf('table.pdf', pages=4)
print(pdfData)
```

```python
pdfData = read_pdf('table.pdf', pages='all')
pdfData
```

```python
tabula.convert_into("table.pdf","output.csv", output_format="csv", pages=4)
```

```python
pdfData = read_pdf('table.pdf', pages=4, lattice=True)
pdfData
```

```python
pdfData = read_pdf('table.pdf', pages=4, stream=True)
pdfData
```

```python
pdfData = read_pdf('table.pdf', pages=4, stream=True)
print(tabulate.tabulate(pdfData))
```

```python
tableData = camelot.read_pdf('table.pdf', multiple_tables='True', process_background='True', pages='4,5')
tableData
```
