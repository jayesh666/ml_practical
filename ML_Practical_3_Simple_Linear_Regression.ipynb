```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv('ML_P_3.csv')
x = d.iloc[:, 0]
y = d.iloc[:, 1]
```

```python
plt.xlabel("Year")
plt.ylabel("Salary")
plt.scatter(x,y)
```

```python
m = 0
c = 0
l = 0.0001
e = 1000

n = float(len(x))

for i in range(e):
    y_pre = m * x + c
    d_m = (-2 / n) * sum (x * (y - y_pre))
    d_c = (-2 / n) * sum (y - y_pre)
    m = m - l * d_m
    c = c - l * d_c
    print("m {}, c {}, iteration{}".format(m,c,i))
    
print("The Slope is:", m)
print("The Intercept is:", c)
```

```python
y_pre = m * x + c

plt.scatter(x, y)
plt.plot(x, y_pre, color="r")
```
